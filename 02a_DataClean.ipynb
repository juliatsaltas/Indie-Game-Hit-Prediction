```python
#[print(k,v) (for i in steam['metacritic'] if type(i) == str) for k,v in i.items()]
```

```python
steam['metacritic_score'] = 0
count=0

for i in steam['metacritic']:

    if type(i) == str:
        #print(i + ' = row contents')
        j = ast.literal_eval(i) # convert string to dictionary
        for k,v in j.items(): # read through key, value pairs
            if k == 'score':
                v_int = int(v) # convert value from string to int
                steam['metacritic_score'][count] = '{0:.3g}'.format(v_int/100) # save score only
            else:
                pass # skip non 'score '
    else:
        steam['metacritic_score'][count] = 0 # assign value of 0 if no metacritic score
        
    count += 1

steam['metacritic_score'].head(10)
```

```python
# drop metacritic column
steam.drop(columns=['metacritic'], inplace=True)
```

```python
# Convert dtype to float
steam['metacritic_score'] = steam['metacritic_score'].astype(float)
```
